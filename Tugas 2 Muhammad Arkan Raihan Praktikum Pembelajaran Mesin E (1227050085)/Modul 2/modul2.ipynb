```python
# %% 
import pandas as pd 

# Load dataset (ubah nama file jika perlu)
data = pd.read_csv("coffee_beans_sales.csv")  

# Menampilkan 10 baris pertama
display(data.head(10))
```

```python
# %%
# Filter data untuk menampilkan hanya kolom tertentu
filtered_data = data.filter(["City", "Category", "Product", "Quantity"])
display(filtered_data.head(10))
```

```python
# %%
# Sorting berdasarkan Quantity tertinggi
data.sort_values("Quantity", axis=0, ascending=False, inplace=True, na_position="last")
display(data.head())
```

```python
# %%
# Group Total Sales by City
grp1 = data.groupby('City')
result = grp1['Final Sales'].sum()
print(result)
```

```python
# %%
# Produk dengan total penjualan terbanyak
top_products = data.groupby("Product")["Quantity"].sum().sort_values(ascending=False)
print(top_products)
```

```python
# %%
# Menghitung rata-rata harga per unit
average_price = data["Unit Price"].mean()
print(f"Rata-rata harga produk: {average_price}")
```

```python
# %%
# Total jumlah diskon yang diberikan
total_discount = data["Discount_Amount"].sum()
print(f"Total diskon yang diberikan: {total_discount}")
```
